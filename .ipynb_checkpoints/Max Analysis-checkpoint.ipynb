```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np
```

```python
# Read in 2021 weekday dataframe

csv_path = "C:/Users/maxgl/Project1-Repo/clean data/weekday_2021_data.csv"
weekday_2021_df = pd.read_csv(csv_path,low_memory=False)
weekday_2021_df.head()
```

```python
# Read in 2022 weekday dataframe

csv_path = "C:/Users/maxgl/Project1-Repo/clean data/weekday_2022_data.csv"
weekday_2022_df = pd.read_csv(csv_path)
weekday_2022_df.head()
```

```python
# Creating Pie Chart to show casual vs member for 2021 rides from 6:00 AM to 6:00 PM 

pie_data = weekday_2021_df["Member type"].value_counts()
pie_df = pd.DataFrame(pie_data)
pie_chart = pie_df.plot(kind='pie', autopct="%1.1f%%", subplots=True, legend=False)
plt.title("Casual vs Member Riders 2021")
plt.show()

# Save the figure
plt.savefig("Bike Share Analysis Images/2021_casual_vs_member.png")
```

```python
# Creating Pie Chart to show casual vs member for 2022 rides from 6:00 AM to 6:00 PM 

pie_data = weekday_2022_df["Member type"].value_counts()
pie_df = pd.DataFrame(pie_data)
pie_chart = pie_df.plot(kind='pie', autopct="%1.1f%%", subplots=True, legend=False)
plt.title("Casual vs Member Riders 2022")
plt.show()

# Save the figure
plt.savefig("Bike Share Analysis Images/2022_casual_vs_member.png")
```

```python
# Now making two new dataframes breaking up 2021 based on vaccination rates in DC

# Creating dataframe for January-April 2021

no_vaccine_df = weekday_2021_df.loc[
    (weekday_2021_df["Month"] == "Jan") |
    (weekday_2021_df["Month"] == "Feb") |
    (weekday_2021_df["Month"] == "Mar") |
    (weekday_2021_df["Month"] == "Apr")                                 
]

# Creating dataframe for May-December 2021

vaccine_df = weekday_2021_df.loc[
    (weekday_2021_df["Month"] == "May") |
    (weekday_2021_df["Month"] == "Jun") |
    (weekday_2021_df["Month"] == "Jul") |
    (weekday_2021_df["Month"] == "Aug") |
    (weekday_2021_df["Month"] == "Sep") |
    (weekday_2021_df["Month"] == "Oct") |
    (weekday_2021_df["Month"] == "Nov") |
    (weekday_2021_df["Month"] == "Dec") 
]

# Display vaccine dataframe

vaccine_df.head()
```

```python
# Display non vaccine dataframe

no_vaccine_df.head()
```

```python
# Create Pie Charts for the new dataframes

# Using data from early 2021 first

pie_data = no_vaccine_df["Member type"].value_counts()
pie_df = pd.DataFrame(pie_data)
pie_chart = pie_df.plot(kind='pie', autopct="%1.1f%%", subplots=True, legend=False)
plt.title("Casual vs Member Riders January-April 2021")
plt.show()

# Save the figure 
plt.savefig("Bike Share Analysis Images/Jan_to_April_2021_casual_vs_member.png")
```

```python
# Now creating the chart for late 2021

pie_data = vaccine_df["Member type"].value_counts()
pie_df = pd.DataFrame(pie_data)
pie_chart = pie_df.plot(kind='pie', autopct="%1.1f%%", subplots=True, legend=False)
plt.title("Casual vs Member Riders May-December 2021")
plt.show()

# Save the figure 
plt.savefig("Bike Share Analysis Images/May_to_December_2021_casual_vs_member.png")
```

```python
# Looking into durations. First looking into whether there is a difference in average duration for members vs casual.

# 2021 data

# Calculate the mean duration for each member type
mean_duration_by_member_type = weekday_2021_df.groupby("Member type")["Duration"].mean()

# Plot the bar chart
mean_duration_by_member_type.plot(kind='bar')
plt.xlabel("Member Type")
plt.ylabel("Average Duration (seconds)")
plt.title("2021 Average Duration of Ride by Member Type")
plt.show()
```

```python
# 2022 data

# Calculate the mean duration for each member type
mean_duration_by_member_type = weekday_2022_df.groupby("Member type")["Duration"].mean()

# Plot the bar chart
mean_duration_by_member_type.plot(kind='bar')
plt.xlabel("Member Type")
plt.ylabel("Average Duration (seconds)")
plt.title("2022 Average Duration of Ride by Member Type")
plt.show()
```

```python
# Calculate the mean duration for each member type in 2021
mean_duration_2021 = weekday_2021_df.groupby("Member type")["Duration"].mean()

# Calculate the mean duration for each member type in 2022
mean_duration_2022 = weekday_2022_df.groupby("Member type")["Duration"].mean()

# Set the width of the bars
bar_width = 0.2

# Set the x positions of the bars
r1 = np.arange(len(mean_duration_2021))
r2 = [x + bar_width for x in r1]

# Plot the bar chart
plt.bar(r1, mean_duration_2021, color='b', alpha=0.5, width=bar_width, label='2021')
plt.bar(r2, mean_duration_2022, color='r', alpha=0.5, width=bar_width, label='2022')

# Set the x-axis labels
plt.xlabel("Member Type")

# Set the y-axis label
plt.ylabel("Average Duration (seconds)")

# Set the title of the plot
plt.title("Average Duration of Ride by Member Type")

# Set the x-axis tick labels
plt.xticks([r + bar_width/2 for r in range(len(mean_duration_2021))], mean_duration_2021.index)

# Add a legend
plt.legend()

# Show the plot
plt.show()

# Save the figure 
plt.savefig("Bike Share Analysis Images/Average_Duration_by_Member_Type.png")
```
