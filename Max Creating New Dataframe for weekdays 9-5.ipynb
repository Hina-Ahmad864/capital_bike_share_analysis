```python
import pandas as pd
```

```python
# Read in 2021 dataframe
csv_path = "C:/Users/maxgl/Project1-Repo/clean data/all_2021_data.csv"
all_2021_df = pd.read_csv(csv_path)
all_2021_df.head()
```

```python
# Read in 2022 dataframe
csv_path = "C:/Users/maxgl/Project1-Repo/clean data/all_2022_data.csv"
all_2022_df = pd.read_csv(csv_path)
all_2022_df.head()
```

```python
# Sorting 2021 into a dataframe with only weekday rides from 6:00 AM to 6:00 PM

# Convert "Start Time" and "End Time" columns to datetime format
all_2021_df["Start Time"] = pd.to_datetime(all_2021_df["Start Time"], format="%I:%M %p")
all_2021_df["End Time"] = pd.to_datetime(all_2021_df["End Time"], format="%I:%M %p")

# Filter the dataframe for weekdays from 6 am to 6 pm
weekday_2021_df = all_2021_df.loc[
    (all_2021_df["Day Type"] == "Weekday") &
    (all_2021_df["Start Time"].dt.time >= pd.to_datetime("06:00").time()) &
    (all_2021_df["End Time"].dt.time <= pd.to_datetime("18:00").time())
].copy()  # Create a copy of the filtered dataframe

# Format the time columns as strings using .loc and df.columns
weekday_2021_df.loc[:, "Start Time"] = weekday_2021_df.loc[:, "Start Time"].dt.strftime("%I:%M %p")
weekday_2021_df.loc[:, "End Time"] = weekday_2021_df.loc[:, "End Time"].dt.strftime("%I:%M %p")

# Display the new dataframe
weekday_2021_df.head()
```

```python
# Run through same process for 2022 data

# Convert "Start Time" and "End Time" columns to datetime format
all_2022_df["Start Time"] = pd.to_datetime(all_2022_df["Start Time"], format="%I:%M %p")
all_2022_df["End Time"] = pd.to_datetime(all_2022_df["End Time"], format="%I:%M %p")

# Filter the dataframe for weekdays from 6 am to 6 pm
weekday_2022_df = all_2022_df.loc[
    (all_2022_df["Day Type"] == "Weekday") &
    (all_2022_df["Start Time"].dt.time >= pd.to_datetime("06:00").time()) &
    (all_2022_df["End Time"].dt.time <= pd.to_datetime("18:00").time())
].copy()  # Create a copy of the filtered dataframe

# Format the time columns as strings using .loc and df.columns
weekday_2022_df.loc[:, "Start Time"] = weekday_2022_df.loc[:, "Start Time"].dt.strftime("%I:%M %p")
weekday_2022_df.loc[:, "End Time"] = weekday_2022_df.loc[:, "End Time"].dt.strftime("%I:%M %p")

# Display the new dataframe
weekday_2022_df.head()
```

```python
# Save the 2021 dataframe as a CSV file
weekday_2021_df.to_csv("weekday_2021_data.csv", index=False)
```
